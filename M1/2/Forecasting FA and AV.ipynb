### Perform Forecasting By converting Time Series to Supervised Learning

```python
#Read Data

import pandas as pd
fa = pd.read_excel('Study Data.xlsx', sheet_name= 'Main Research Subjects', usecols= ['السنوات', 'المحاسبة المالية'], parse_dates=['السنوات'])
fa.set_index('السنوات', inplace=True)

#Framing data to enable forecasting time series as supervised learning - Linear Regression & SVM are utilized
shifted_fa = fa.copy()
shifted_fa['y-1'] = shifted_fa['المحاسبة المالية'].shift(1)
shifted_fa['y+1'] = shifted_fa['المحاسبة المالية'].shift(-1)
shifted_fa.dropna(how='any', inplace=True)

#Splitting data for training and testing
X = shifted_fa.iloc[:, : -1].values
y = shifted_fa.iloc[:, -1:].values
X_train, X_test, y_train, y_test = X[:-2], X[-2:], y[:-2], y[-2:]
```

```python
#Apply Linear Regression Prediction (Forecasting)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.predict(X_test)
print('predicted', lr.predict(X_test))
print('Actual', y_test.ravel())
print('LR R^2 = ', lr.score(X_test, y_test))
#Financial Accounting Forecasting for 2018
print('Financial Accounting for 2018 utilizing LR = ', lr.predict([[4.0, 3.0]]))
print ('------------------------------------------------------------')



#Apply SVM Prediction (Forecasting)
from sklearn.svm import SVR
svr = SVR(gamma='scale', C=2)
svr.fit(X_train, y_train.ravel())
print('predicted', svr.predict(X_test))
print('Actual', y_test.ravel())
print('svr R^2 = ', svr.score(X_test, y_test.ravel()))
#Financial Accounting Forecasting for 2018
print('Financial Accounting for 2018 utlizing SVM = ', svr.predict([[4.0, 3.0]]))
print ('------------------------------------------------------------')


#Apply Randomforest Prediction (Forecasting)
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=1000, max_depth=2, random_state=42)
rfr.fit(X_train, y_train.ravel())
print('predicted', rfr.predict(X_test))
print('Actual', y_test.ravel())
print('RF R^2 = ', rfr.score(X_test, y_test.ravel()))
#Financial Accounting Forecasting for 2018
print('Financial Accounting for 2018 utilizing RF= ', rfr.predict([[4.0, 3.0]]))
```

### Perform Forecasting Using Time Series ARIMA Model

```python
#Utilizing ARIMA Model
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pyplot as plt
%matplotlib inline

#devide data for train and test
X = fa.values
X_train, X_test= X[:-2], X[-2:]
predictions = []


arima_mod = ARIMA(X_train, order= (1,1,1))
arima_mod_fit = arima_mod.fit()
print('AIC = ', arima_mod_fit.aic)

#predictions
predictions = arima_mod_fit.forecast(steps = 3)[0]
predictions

plt.plot(X_test)
plt.plot(predictions, color = 'red');

print('Financial Accounting for 2018 utilizing ARIMA= ', predictions[2])
```

```python
#Read Data
import pandas as pd
av = pd.read_excel('Study Data.xlsx', sheet_name= 'Main Research Subjects', usecols= ['السنوات', 'المراجعة والتوكيد'], parse_dates=['السنوات'])
av.set_index('السنوات', inplace=True)

#Framing data to enable forecasting time series as supervised learning - Linear Regression & SVM are utilized
shifted_av = av.copy()
shifted_av['y-1'] = shifted_av['المراجعة والتوكيد'].shift(1)
shifted_av['y+1'] = shifted_av['المراجعة والتوكيد'].shift(-1)
shifted_av.dropna(how='any', inplace=True)

#Splitting data for training and testing
X = shifted_av.iloc[:, : -1].values
y = shifted_av.iloc[:, -1:].values
X_train, X_test, y_train, y_test = X[:-2], X[-2:], y[:-2], y[-2:]
```

```python
#Apply Linear Regression Prediction (Forecasting)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.predict(X_test)
print('predicted', lr.predict(X_test))
print('Actual', y_test.ravel())
print('LR R^2 = ', lr.score(X_test, y_test))
#Financial Accounting Forecasting for 2018
print('Financial Accounting for 2018 utlizaing LR= ', lr.predict([[4.0, 3.0]]))
print ('------------------------------------------------------------')



#Apply SVM Prediction (Forecasting)
from sklearn.svm import SVR
svr = SVR(gamma='scale', C=2)
svr.fit(X_train, y_train.ravel())
print('predicted', svr.predict(X_test))
print('Actual', y_test.ravel())
print('svr R^2 = ', svr.score(X_test, y_test.ravel()))
#Financial Accounting Forecasting for 2018
print('Financial Accounting for 2018 utlizaing SVM= ', svr.predict([[4.0, 3.0]]))
print ('------------------------------------------------------------')


#Apply Randomforest Prediction (Forecasting)
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=1000, max_depth=2, random_state=42)
rfr.fit(X_train, y_train.ravel())
print('predicted', rfr.predict(X_test))
print('Actual', y_test.ravel())
print('RF R^2 = ', rfr.score(X_test, y_test.ravel()))
#Financial Accounting Forecasting for 2018
print('Financial Accounting for 2018 utlizaing RF= ', rfr.predict([[4.0, 3.0]]))
```

### Perform Forecasting Using Time Series ARIMA Model

```python
#Utilizing ARIMA Model
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pyplot as plt
%matplotlib inline

#devide data for train and test
X = av.values
X_train, X_test= X[:-2], X[-2:]
X_train = X_train.ravel().astype('float32')
X_test = X_test.ravel().astype('float32')
predictions = []


#Find best combinations between p,d,q
import itertools
import warnings
warnings.filterwarnings('ignore')
p=d=q= range(0,5)
pdq = list(itertools.product(p,d,q))
comb = []
for parm in pdq:
    try:
        arima_mod = ARIMA(X_train, order = parm)
        arima_mod_fit = arima_mod.fit()
        comb.append((parm, arima_mod_fit.aic))
    except:
        continue
        
lowest=[(0,0,0), 100000]
for i, v in enumerate(comb):
    if comb[i][1] < lowest[1]:
        lowest = comb[i]
        
print('Lowest Model parameters are: ', lowest)


#Selecting best model and perform forecasting
arima_mod = ARIMA(X_train, order= lowest[0])
arima_mod_fit = arima_mod.fit()
print('AIC = ', arima_mod_fit.aic)

#predictions
predictions = arima_mod_fit.forecast(steps = 3)[0]
predictions

#plt.plot(X_test, 'b-+')
#plt.plot(predictions, 'r-*');

#print('Auditing & Verification Accounting for 2018 = ', predictions[2])
```

```python
print('Lowest Model parameters are: ', lowest)

plt.plot(X_test, 'b-+')
plt.plot(predictions, 'r-*');

print('Auditing & Verification Accounting for 2018 utlizaing ARIMA= ', predictions[2])
```

##### Stationarity
means the mean, variance and coverience are constant over period of time.

```python
# Checking stationarity through auto correlation model
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(fa)
```

If the figure shows slowly decay, means the series is not stationary. what we see above is a rapid decay which means the series is stationary. I'll even try to make it decreasing faster.

```python
## Converting series to stationary (by making it Integrated of order 1 --> d=1 which is one of the parameters of ARIMA model)
fadiff = fa.diff(periods=1)
fadiff.dropna(how='any', inplace=True)
fadiff.head(3)
```

```python
# Checking stationarity through auto correlation model
plot_acf(fadiff)
```

```python
# let's check the figure
fadiff.plot();
```

##### ARMA Model

```python
from statsmodels.tsa.arima_model import ARMA
from sklearn.metrics import mean_squared_error

#devide data for train and test
X = fa.values
X_train, X_test= X[:-2], X[-2:]
predictions = []

#Find best combinations between p,d,q
import warnings
warnings.filterwarnings('ignore')
p=q= range(0,5)
pq = list(itertools.product(p,q))
comb = []
for parm in pq:
    try:
        arma_mod = ARMA(X_train, order = parm)
        arma_mod_fit = arma_mod.fit()
        comb.append((parm, arma_mod_fit.aic))
    except:
        continue
        
lowest=[(0,0), 100000]
for i, v in enumerate(comb):
    if comb[i][1] < lowest[1]:
        lowest = comb[i]


# To perform ARMA, we need to specify its 2 parameters:
# p,q --> AR (number of shifts of data), MA (number of shifts of differences)

arma_mod = ARMA(X_train, order= lowest[0])
#arma_mod = ARMA(X_train, order= (1,1))
arma_mod_fit = arma_mod.fit()
print('ARMA AIC', arma_mod_fit.aic)

# Forecast last 3 years
predictions = arma_mod_fit.forecast(steps = 3)[0]
print('Financial Accounting for 2018 utilizing ARMA= ', predictions[2])

# Plotting Actual against Prediction
plt.xticks(range(0,3), range(2016, 2019))
plt.plot(X_test)
plt.plot(predictions, color = 'red');
```
